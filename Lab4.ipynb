## a)

Se tiene que para retornar los dos diccionarios, al menos m-n soluciones serán básicas y el identificar la base a la que pertenecen va a ser el principal reto. Por ello, se consideran todas las posibles soluciones y los casos en que la matriz se define o no de manera completa.

```
def SolucionesBasicas(A,b):
  # Diccionario de visualización
  dic={'Solución': [],'Base': []}
  
  rango=np.linalg.matrix_rank(A)
  zcA=np.size(A[1,:])
  A2=A[0:rango,0:rango]
  A3=A.copy();

  # Creacion de espacios de iteración
  z=np.linspace(0,zcA-1,zcA)
  r=np.linspace(0,rango-1,rango)
  
  # Nuevo vector de soluciones de acuerdo a la nueva matriz
  b2=b[:rango]

  #Recorrido en todas las combinaciones posibles de elementos de la nueva matriz
  for i in combinations(z,rango):
    
    # Creacion de la nueva matriz
    for j in r:
      A2[:,int(j)]=A3[:rango,int(i[int(j)])].copy()

    # Caso en el que la nueva matriz no es de rango completo 
    if np.linalg.det(A2)==0:
      dic1=SolucionesBasicas(A2,b2) 
      dic['Solución'].extend(dic1['Solución'])
      dic['Base'].extend(dic1['Base'])  
    # Caso en el que la nueva matriz es de rango completo  
    else:
      k=np.linalg.solve(A2,b2)
      dic['Solución'].append(k)
      dic['Base'].append(A2.copy().transpose()) 
  return dic
```

Se implementó una función que dada una matriz de coeficientes $A$, un vector de constantes $b$ y una función de costo $f$ que corresponden a una problema de optimización en su forma estándar, devuelve un diccionario que contiene las siguientes entradas: 

- Solución: Soluciones básicas del problema. 
- Base: Matriz base con la que se obtuvo la solución del problema.  
- Tipo: Tipo de solución habiendo 2 categorias, *solución básica factible* y *solución básica no factible*. 
- Valor: Valor de la función de costo en la solución básica encontrada.

**Entradas:** 
- $A$: Matriz de coeficientes asociados a las variables de optimización.  
- $b$: Vector de constantes 
- $f$: Función de costo asociada al problema. 

**Salidas:**
- dict: Diccionario conteniendo los items previamente descritos. 

Para poder emplear esta función, se creó otra función llamada negative_check

```
def negative_check(L):  
  check = False 
  for i in L: 
    if (i < 0): 
      check = True
      break 
  return check
```

```
import numpy as np

def sol_est_problem(A,b,f): 

  # Se convierten los arreglos de entrada en numpy arrays.  
  A = np.array(A)
  b = np.array(b).T
  f = np.array(f)    

  # Parámetros del sistema lineal
  m,n = np.shape(A)   

  # Parametros de importancia 
  limite = f.shape[0] # Número de variables reales del problema 
  # Diccionario de visualización
  data = {'Solucion':[],'Base':[], 'Tipo':[], 'Valor':[]} 
  # Todas las soluciones básicas
  solutions = SolucionesBasicas(A,b)

  # Caso en que hay una única solución
  if n == m:  
    data['Solucion'] = solutions['Solución']
    data['Base'] = solutions['Base']
    data['Tipo'] = 'Única solución' 
    S = S[0:limite] 
    try:
      value = np.dot(S,f) # Evaluacion de la solucion
    except: 
      value = None
    data['Valor'].append(value)

  # Caso de interés, n_a>m_b donde n_a-m_b = 2.
  elif n >m :  
    data['Base'] = solutions['Base']
    for S in solutions['Solución']:
    # Clasificacion de las soluciones
      try: 
        check = negative_check(S)
      except: 
        check = True

      if check == True:
        value = None 
        data['Valor'].append(value) 
        data['Tipo'].append('Solución básica no factible')
        try:
          S = S[0:limite]
        except: 
          S = None
        data['Solucion'].append(S) 
      else: 
        S = S[0:limite]
        value = np.dot(S,f) # Evaluación de la solución
        data['Valor'].append(value) 
        data['Tipo'].append('Solución básica factible') 
        data['Solucion'].append(S) 

  # Caso donde no hay solución. 
  elif m > n:  
    data['Solucion'] = None
    data['Base'] = np.array(A)
    data['Tipo'] = None
    data['Valor'] = None
  
  return data
```

```
from IPython.display import display, HTML 
import pandas as pd 
import numpy as np


A =[[20,18,15,1,0,0],
    [15,12,10,0,1,0],
    [200,220,250,0,0,1]] 
b = [60,75,2000]  
f = [-600,-550,-500] 

s = sol_est_problem(A,b,f)
p = pd.DataFrame(data=s)
display(p)
```
